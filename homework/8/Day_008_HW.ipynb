```python
app_train['AMT_INCOME_TOTAL']
```

```python
print('收入平均值：', app_train['AMT_INCOME_TOTAL'].mean())
print('收入標準差：', app_train['AMT_INCOME_TOTAL'].std())
print('最高收入：', app_train['AMT_INCOME_TOTAL'].max())
print('最低收入：', app_train['AMT_INCOME_TOTAL'].min())
```

```python
print('整體\n', app_train['AMT_INCOME_TOTAL'].describe())
```

```python
plt.figure(figsize = (10,10))
plt.title('Income about customer')
plt.xlabel('Income')
plt.ylabel('Customer')
plt.yticks(range(0, app_train['AMT_INCOME_TOTAL'].shape[0], 10000))
plt.hist(app_train['AMT_INCOME_TOTAL'], color = 'blue', edgecolor = 'white', label = 'AMT_INCOME_TOTAL')
plt.grid(linestyle = '-')
plt.legend()
plt.show()
```
